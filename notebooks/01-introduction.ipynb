## üèÅ What is Data Visualization?

Data visualization is the process of **turning raw data into visual insights** using charts, graphs, and plots.  

It helps:
- Identify **patterns, trends, and outliers**.
- Communicate findings to both technical and non-technical audiences.
- Support **data-driven decision making**.

---



---

## üéØ Why Does It Matter?

Raw data can be overwhelming üëá

| Name | Sales | Month |
|------|-------|-------|
| A    | 120   | Jan   |
| B    | 90    | Jan   |
| A    | 150   | Feb   |
| B    | 130   | Feb   |

But when visualized: 

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb"]
sales_A = [120, 150]
sales_B = [90, 130]

plt.plot(months, sales_A, marker='o', label="A")
plt.plot(months, sales_B, marker='o', label="B")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

## üõ† Python Libraries for Visualization

We‚Äôll use three key libraries:

1. **Matplotlib** ‚Üí low-level, very flexible plotting library.  
2. **Seaborn** ‚Üí built on Matplotlib, provides beautiful statistical plots.  
3. **Plotly** ‚Üí interactive visualizations (we‚Äôll explore later).  

```python
# Quick preview of each library

import seaborn as sns
import pandas as pd

# Simple dataset
tips = sns.load_dataset("tips")

# Matplotlib
tips["total_bill"].plot(kind="hist", bins=20, title="Matplotlib Example")
plt.show()

# Seaborn
sns.histplot(data=tips, x="total_bill", bins=20)
plt.title("Seaborn Example")
plt.show()

# Plotly
import plotly.express as px
fig = px.histogram(tips, x="total_bill", nbins=20, title="Plotly Example")
fig.show()
```
