In this notebook, weâ€™ll explore three of the **most commonly used plot types** in data visualization:

* **Line Plots** â†’ Great for trends over time
* **Bar Plots** â†’ Best for comparing categories
* **Scatter Plots** â†’ Useful for showing relationships between two variables

Weâ€™ll use **Matplotlib** and **Seabornâ€™s `tips` dataset** to demonstrate each type.

---

## ðŸ“Œ 1. Setup & Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load sample dataset
tips = sns.load_dataset("tips")
tips.head()
```


---

## ðŸ“ˆ 2. Line Plots â€” Trends Over Time

A **line plot** is useful when you want to see changes over a continuous variable (e.g., time, sequence, temperature).

```python
# Example: Plotting total_bill trend by index
plt.figure(figsize=(8,5))
plt.plot(tips["total_bill"], color="blue", marker="o", linestyle="-")
plt.title("Line Plot of Total Bill")
plt.xlabel("Index")
plt.ylabel("Total Bill ($)")
plt.show()
```


ðŸ”‘ **Key Notes:**

* `plt.plot()` â†’ default line chart
* `marker="o"` â†’ adds dots to points
* Line plots work best for continuous or ordered data

ðŸ‘‰ Example usage: stock prices, monthly sales, temperature change

---

## ðŸ“Š 3. Bar Plots â€” Comparing Categories

Bar plots show comparisons across **discrete categories**.

```python
# Example: Average tip per day
avg_tips = tips.groupby("day")["tip"].mean()

plt.figure(figsize=(7,5))
plt.bar(avg_tips.index, avg_tips.values, color="orange")
plt.title("Average Tip by Day")
plt.xlabel("Day of Week")
plt.ylabel("Average Tip ($)")
plt.show()
```

ðŸ”‘ **Key Notes:**

* `plt.bar(categories, values)` for vertical bars
* `plt.barh()` for horizontal bars
* Perfect for categorical comparisons

ðŸ‘‰ Example usage: sales per region, votes per candidate, website traffic per device

---

## âšª 4. Scatter Plots â€” Relationship Between Two Variables

Scatter plots are best when exploring **correlation** between variables.

```python
# Example: Tip vs. Total Bill
plt.figure(figsize=(7,5))
plt.scatter(tips["total_bill"], tips["tip"], alpha=0.7, color="green")
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```


ðŸ”‘ **Key Notes:**

* `plt.scatter(x, y)` creates scatter plots
* `alpha` adds transparency to avoid overlap
* Helpful for spotting **trends, outliers, clusters**

ðŸ‘‰ Example usage: income vs. spending, hours studied vs. exam scores

---

## ðŸŽ¯ 5. Comparison of Plot Types

| Plot Type    | Best For                           |

## ðŸ”¹ . Line Plots

```python
import numpy as np
# Sample data
x = np.arange(0, 10, 1)
y = x ** 2

plt.figure(figsize=(8,5))
plt.plot(x, y, color="blue", marker="o", linestyle="-", label="y = x^2")
plt.title("Line Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

ðŸ‘‰ **Explanation**:  
- `plt.plot()` creates the line graph.  
- We customized with `color`, `marker`, and `linestyle`.  
- `plt.legend()` helps clarify multiple lines.  
- Line plots are best when **tracking changes or trends**.

---

## ðŸ”¹. Bar Plots

```python
# Categories and values
categories = ["A", "B", "C", "D"]
values = [4, 7, 1, 8]

plt.figure(figsize=(7,5))
plt.bar(categories, values, color="orange")
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```


ðŸ‘‰ **Explanation**:  
- `plt.bar()` creates vertical bars.  
- Useful for comparing discrete categories.  
- Bar height shows the **magnitude of values**.


### Horizontal Bar Plot

```python

plt.figure(figsize=(7,5))
plt.barh(categories, values, color="green")
plt.title("Horizontal Bar Plot Example")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

ðŸ‘‰ **Note**: Horizontal bars (`plt.barh`) are often better when category names are long.

---


## ðŸ”¹. Scatter Plots

```python
# Random scatter data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(7,5))
plt.scatter(x, y, color="purple", alpha=0.7, s=100, edgecolor="black")
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```


ðŸ‘‰ **Explanation**:  
- `plt.scatter()` plots individual points.  
- Parameters:  
  - `alpha` â†’ transparency  
  - `s` â†’ size of markers  
  - `edgecolor` â†’ outline around dots  
- Scatter plots are ideal for spotting **correlations or clusters**.

---

## ðŸ”¹ . Comparing Multiple Lines

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(8,5))
plt.plot(x, np.sin(x), label="sin(x)", color="red")
plt.plot(x, np.cos(x), label="cos(x)", color="blue")
plt.title("Multiple Lines Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```


ðŸ‘‰ Multiple lines in the same chart make it easy to compare trends or functions.
