# üìä Matplotlib Basics

Matplotlib is the **foundation of data visualization in Python**.  
It gives you low-level control to build almost any kind of static, animated, or interactive plot.

In this notebook, we‚Äôll cover:
1. Installing & Importing Matplotlib  
2. Basic Plotting (`plt.plot`)  
3. Adding Labels & Titles  
4. Adjusting Colors & Styles  
5. Saving Plots

---


`````python
# 1. Importing Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Example dataset
x = np.linspace(0, 10, 100)   # 100 values between 0 and 10
y = np.sin(x)
`````

---


## üìà Basic Line Plot

The simplest plot in Matplotlib is a **line plot** using `plt.plot()`.

`````python
plt.plot(x, y)
plt.show()
`````

---

## üè∑Ô∏è Adding Labels & Titles

We can add:
- **X-axis label** ‚Üí `plt.xlabel()`  
- **Y-axis label** ‚Üí `plt.ylabel()`  
- **Title** ‚Üí `plt.title()`  

`````python
plt.plot(x, y)
plt.xlabel("X Axis (Time)")
plt.ylabel("Y Axis (Amplitude)")
plt.title("Sine Wave Example")
plt.show()
`````


---

## üé® Changing Colors, Line Styles & Markers

Matplotlib lets you customize the look of your plots.

Common options:
- `color='red'` or shorthand `color='r'`
- `linestyle='--'` (dashed), `':'` (dotted), `'-.'` (dash-dot)
- `marker='o'` (circle), `'*'` (star), `'s'` (square)

`````python
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Customized Sine Wave")
plt.show()
`````


---

## üìë Multiple Plots in One Figure

We can plot multiple lines on the same chart by calling `plt.plot()` multiple times before `plt.show()`.

`````python
y2 = np.cos(x)

plt.plot(x, y, label="Sine Wave")
plt.plot(x, y2, label="Cosine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Sine vs Cosine")
plt.legend()
plt.show()
`````


---

## üíæ Saving Figures

You can save a plot to a file (PNG, JPG, PDF, etc.) using:

```python
plt.savefig("plot.png")
````

`````python
plt.plot(x, y, label="Sine Wave")
plt.title("Saving a Plot Example")
plt.legend()
plt.savefig("sine_plot.png", dpi=300)   # saves high-quality PNG
plt.show()
`````
