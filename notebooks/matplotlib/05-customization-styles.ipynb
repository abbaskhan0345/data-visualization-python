## ðŸ“Œ 1. Adding Titles, Labels, and Legends

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,5))
plt.plot(x, y1, label="sin(x)", color="blue")
plt.plot(x, y2, label="cos(x)", color="red")

plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis (radians)")
plt.ylabel("Y-axis (value)")
plt.legend()

plt.show()
```


---


## ðŸ“Œ 2. Colors, Line Styles, and Markers
Matplotlib lets you customize almost every visual element.

```python

plt.figure(figsize=(8,5))
plt.plot(x, y1, color="green", linestyle="--", marker="o", label="sin(x)")
plt.plot(x, y2, color="purple", linestyle="-.", marker="s", label="cos(x)")

plt.title("Custom Colors, Styles & Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```


---

## ðŸ“Œ 3. Grid and Axis Customization

```python
plt.figure(figsize=(8,5))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

plt.title("With Grid and Axis Customization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Enable grid
plt.grid(True, linestyle="--", alpha=0.6)

# Axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

plt.show()
```

---

## ðŸ“Œ 4. Using Built-in Stylesheets
Matplotlib comes with **predefined styles**.

```python
print(plt.style.available)  # list of available styles
```

```python
# Example: applying different styles
plt.style.use("ggplot")

plt.figure(figsize=(8,5))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Styled with ggplot")
plt.legend()
plt.show()
```

```python
# Another example
plt.style.use("seaborn-v0_8-dark-palette")

plt.figure(figsize=(8,5))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Styled with seaborn-dark-palette")
plt.legend()
plt.show()
```

```python

# Reset style
plt.style.use("default")
```

---


## ðŸ“Œ 5. Annotations
Annotations allow us to **highlight specific points** in a plot.

```python
plt.figure(figsize=(8,5))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

plt.title("Adding Annotations Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

```python

# Annotate peak
peak_x = np.pi/2
peak_y = np.sin(peak_x)
plt.annotate("Peak of sin(x)", xy=(peak_x, peak_y), xytext=(2, 1),
             arrowprops=dict(facecolor="black", arrowstyle="->"))

plt.show()
```


---

## ðŸ“Œ 6. Saving Figures
We can export plots in **PNG, JPG, SVG, PDF** and more.

```python
plt.figure(figsize=(8,5))
plt.plot(x, y1, label="sin(x)")
plt.title("Saving Figure Example")
plt.legend()

# Save as PNG
plt.savefig("sine_plot.png", dpi=300)

plt.show()
```
