## ðŸ“ˆ Histograms â€“ Understanding Distributions

A **histogram** groups data into intervals (bins) and counts how many values fall into each bin.  
This is useful for seeing **data distribution** patterns.

```python
# Histogram of petal length
plt.hist(iris["petal length (cm)"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```


---

### ðŸ”§ Customizing Histograms
- `bins` â†’ number of intervals  
- `color` â†’ bar fill  
- `edgecolor` â†’ outline of bars  
- `alpha` â†’ transparency

```python
plt.hist(iris["sepal length (cm)"], bins=15, color="orange", alpha=0.7, edgecolor="black")
plt.title("Histogram of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Count")
plt.show()
```


---

## ðŸ“¦ Boxplots â€“ Spread & Outliers

A **boxplot** shows:  
- **Median (middle line)**  
- **Q1 & Q3 (box edges â†’ interquartile range, IQR)**  
- **Whiskers (min & max within 1.5Ã—IQR)**  
- **Outliers (points beyond whiskers)**  

```python
# Single variable boxplot
plt.boxplot(iris["petal length (cm)"])
plt.title("Boxplot of Petal Length")
plt.ylabel("Petal Length (cm)")
plt.show()
```


---

### ðŸ“Š Multiple Boxplots â€“ Comparing Groups

Boxplots are powerful for comparing distributions across **categories**.

```python
# Grouped boxplot by species
species = iris["type"].unique()
data = [iris[iris["type"] == sp]["petal length (cm)"] for sp in species]

plt.boxplot(data, labels=species)
plt.title("Boxplot of Petal Length by Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

---

```python

# Sample dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=200)  # Normally distributed data
```



---


## ðŸ”¹ Histograms
A **histogram** groups data into bins and shows the frequency (or probability density) of values.  

```python
plt.figure(figsize=(7,5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Data", fontsize=14)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```


---

## ðŸ”¹ Boxplots
A **boxplot** summarizes data using the median, quartiles (Q1, Q3), and potential outliers.  


```python
plt.figure(figsize=(7,3))
sns.boxplot(x=data, color="lightcoral")
plt.title("Boxplot of Data", fontsize=14)
plt.show()
```

---

## ðŸ”¹ Side-by-Side Comparison
To better understand the **distribution vs summary**, we plot a **histogram and boxplot together** for the same variable.

```python
fig, axes = plt.subplots(2, 1, figsize=(8,6), sharex=True)

# Histogram
axes[0].hist(data, bins=15, color='skyblue', edgecolor='black')
axes[0].set_title("Histogram", fontsize=12)

# Boxplot
sns.boxplot(x=data, color="lightcoral", ax=axes[1])
axes[1].set_title("Boxplot", fontsize=12)

plt.suptitle("Histogram vs Boxplot", fontsize=14)
plt.tight_layout()
plt.show()
```


---

## ðŸ”¹ Multiple Distributions
We can compare histograms and boxplots for different groups. Letâ€™s create two datasets and compare.  

```python
# Two groups
group1 = np.random.normal(60, 8, 200)
group2 = np.random.normal(45, 12, 200)

# Side-by-side plots
fig, axes = plt.subplots(2, 2, figsize=(12,8))

# Histograms
axes[0,0].hist(group1, bins=15, color='skyblue', edgecolor='black')
axes[0,0].set_title("Histogram - Group 1")
axes[0,1].hist(group2, bins=15, color='lightgreen', edgecolor='black')
axes[0,1].set_title("Histogram - Group 2")

# Boxplots
sns.boxplot(x=group1, color="lightcoral", ax=axes[1,0])
axes[1,0].set_title("Boxplot - Group 1")
sns.boxplot(x=group2, color="orange", ax=axes[1,1])
axes[1,1].set_title("Boxplot - Group 2")

plt.suptitle("Histograms & Boxplots for Two Groups", fontsize=16)
plt.tight_layout()
plt.show()
```


## âœ… Summary

- **Histograms**:  
  - Show frequency distribution of a variable.  
  - Controlled by **bins**, **colors**, and **styles**.  

- **Boxplots**:  
  - Show **median, quartiles, spread, and outliers**.  
  - Useful for comparing distributions across categories.  

ðŸ“Œ With histograms, we **understand shape** of data; with boxplots, we **summarize spread and outliers**.  

- **Histograms** â†’ Show *distribution shape* (normal, skewed, multimodal, etc.).  
- **Boxplots** â†’ Highlight *median, spread, and outliers*.  
- **Together** â†’ Provide both *visual intuition* and *statistical summary*.
