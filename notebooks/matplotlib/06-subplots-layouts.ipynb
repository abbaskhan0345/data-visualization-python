In real-world data visualization, we often want to **compare multiple plots side by side** or organize them into a **grid layout** for better storytelling.  
Matplotlib makes this possible using **`plt.subplots()`** and related functions.

## 🔹 1. Why Subplots?

Instead of generating multiple figures separately, we can organize them together:

✅ Easier comparison.  
✅ Cleaner and more compact presentation.  
✅ Often used in dashboards, reports, or EDA.

---

## 🔹 2. Creating Subplots with `plt.subplots()`

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create 1 row, 2 columns
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Plot on first subplot
ax[0].plot(x, y1, color="blue", label="sin(x)")
ax[0].set_title("Sine Function")
ax[0].legend()

# Plot on second subplot
ax[1].plot(x, y2, color="red", label="cos(x)")
ax[1].set_title("Cosine Function")
ax[1].legend()

plt.suptitle("📊 Side-by-Side Subplots Example", fontsize=14)
plt.show()
```


---

## 🔹 3. Multi-Row Layouts


```python
fig, ax = plt.subplots(2, 1, figsize=(6, 6))

ax[0].plot(x, y1, color="green")
ax[0].set_title("Sine Wave")

ax[1].plot(x, y2, color="orange")
ax[1].set_title("Cosine Wave")

plt.tight_layout()
plt.suptitle("Multiple Rows of Subplots", fontsize=14, y=1.02)
plt.show()
```


---


## 🔹 4. Grid Layouts

```python

fig, ax = plt.subplots(2, 2, figsize=(10, 6))

ax[0, 0].plot(x, y1, color="blue")
ax[0, 0].set_title("sin(x)")

ax[0, 1].plot(x, y2, color="red")
ax[0, 1].set_title("cos(x)")

ax[1, 0].scatter(x, y1, color="purple")
ax[1, 0].set_title("Scatter of sin(x)")

ax[1, 1].bar([1,2,3], [5,7,3], color="cyan")
ax[1, 1].set_title("Bar Example")

plt.tight_layout()
plt.suptitle("2x2 Grid Layout of Subplots", fontsize=14, y=1.05)
plt.show()
```


---


## 🔹 5. Shared Axes

Sometimes, we want all plots to share the same **x-axis** or **y-axis** for consistency.

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 6), sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        ax[i, j].plot(x, np.sin(x + (i+j)), label=f"sin(x+{i+j})")
        ax[i, j].legend()

plt.suptitle("Subplots with Shared Axes", fontsize=14, y=1.02)
plt.show()
```


---


## 🔹 6. Adjusting Layouts

* `plt.tight_layout()` → Automatically adjusts spacing.
* `plt.subplots_adjust()` → Fine-tunes spacing manually.

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 6))

for i in range(2):
    for j in range(2):
        ax[i, j].plot(x, np.sin(x + i + j))

plt.subplots_adjust(wspace=0.3, hspace=0.4)  # Adjust width/height spacing
plt.show()
```


---


## 📝 Summary

* **Subplots** let you organize multiple plots in one figure.
* Use `plt.subplots(rows, cols)` to create grids.
* **Shared axes** help with consistent comparisons.
* Layouts can be adjusted using `tight_layout()` or `subplots_adjust()`.
