# üìä Categorical Plots in Seaborn
**`09-categorical-plots.ipynb`**

Categorical plots are used to **compare categories** rather than continuous distributions.  
Seaborn provides multiple tools to visualize **counts, proportions, and category-wise distributions**.

We‚Äôll use the `tips` dataset.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")
tips.head()
```


---

## 1Ô∏è‚É£ Bar Plot (`barplot`)

```python

sns.barplot(x="sex", y="total_bill", data=tips, estimator=sum, ci=None)
plt.title("Total Bill by Gender")
plt.show()
```


‚úÖ By default, `barplot` shows the **mean** with confidence interval.
Here we set `estimator=sum` to see total bill instead.

---

## 2Ô∏è‚É£ Count Plot (`countplot`)


```python
sns.countplot(x="day", data=tips, hue="sex")
plt.title("Count of Customers by Day and Gender")
plt.show()
```


‚úÖ Shows the **number of observations** per category.

---


## 3Ô∏è‚É£ Box Plot (`boxplot`)


```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Boxplot of Total Bill by Day and Gender")
plt.show()
```

‚úÖ Boxplots display **median, quartiles, and outliers**.

---

## 4Ô∏è‚É£ Violin Plot (`violinplot`)

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```


‚úÖ Combines a **boxplot + KDE distribution**.

---

## 5Ô∏è‚É£ Strip Plot (`stripplot`)


```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("Strip Plot of Total Bill by Day")
plt.show()
```

‚úÖ Shows **individual data points** with some jitter to reduce overlap.

---

## 6Ô∏è‚É£ Swarm Plot (`swarmplot`)

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Swarm Plot of Total Bill by Day and Gender")
plt.show()
```


‚úÖ Like a strip plot but prevents overlap ‚Üí **denser, more readable**.

---

## 7Ô∏è‚É£ Combined Categorical Plots

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None, color="lightgray")
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
plt.title("Violin + Swarm Plot")
plt.show()
```


‚úÖ Overlaying plots gives deeper insights.

---

# üìå Summary

- **Bar Plot** ‚Üí category means (or other stats with `estimator`)  
- **Count Plot** ‚Üí frequency counts  
- **Box Plot** ‚Üí quartiles + outliers  
- **Violin Plot** ‚Üí KDE + box info  
- **Strip/Swarm Plot** ‚Üí individual data points
