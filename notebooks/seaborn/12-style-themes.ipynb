## ðŸ”¹ 1. Matplotlib Styles
Matplotlib provides built-in styles you can easily apply.

```python

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Default style
plt.plot(x, y, label="Default Style")
plt.title("Default Style Example")
plt.legend()
plt.show()
```

```python
# Applying different styles
for style in ['ggplot', 'seaborn', 'fivethirtyeight', 'classic']:
    plt.style.use(style)
    plt.plot(x, y, label=style)
    plt.title(f"Matplotlib Style: {style}")
    plt.legend()
    plt.show()

# Reset to default
plt.style.use('default')
```


---


## ðŸ”¹ 2. Seaborn Themes
Seaborn has built-in themes for overall aesthetics:
- `darkgrid` (default)
- `whitegrid`
- `dark`
- `white`
- `ticks`

```python
import seaborn as sns

# Load dataset
tips = sns.load_dataset("tips")

for theme in ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']:
    sns.set_theme(style=theme)
    sns.scatterplot(x="total_bill", y="tip", data=tips)
    plt.title(f"Seaborn Theme: {theme}")
    plt.show()

# Reset to default
sns.set_theme()
```


---

## ðŸ”¹ 3. Seaborn Color Palettes
Customizing **colors** is another way to improve visual appeal.


```python
# Available color palettes
print(sns.color_palette("Set2"))
sns.set_palette("Set2")

sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Seaborn Palette: Set2")
plt.show()

# Try another palette
sns.set_palette("coolwarm")
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Seaborn Palette: coolwarm")
plt.show()
```

---



## ðŸ”¹ 4. Combining Styles + Palettes
We can mix themes and palettes for ultimate customization.


```python
sns.set_theme(style="whitegrid", palette="muted")

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Custom Style + Palette Example")
plt.show()
```
