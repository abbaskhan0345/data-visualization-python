Seaborn is a **statistical data visualization library** built on top of Matplotlib.  
It provides a high-level interface for creating **beautiful, informative graphics** with minimal code.

---


## ðŸ“Œ Why Seaborn?
- Built on **Matplotlib** but more **aesthetic** out of the box.
- Excellent for **statistical visualization** (distributions, correlations, categorical plots).
- Integrates smoothly with **Pandas DataFrames**.
- Provides built-in **themes and color palettes**.

---

## ðŸ”§ Setup


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load an inbuilt dataset
tips = sns.load_dataset("tips")
tips.head()
```


---


## ðŸŽ¨ Seaborn Themes

Seaborn comes with several **themes** for better styling:

```python
sns.set_style("whitegrid")  # other options: "darkgrid", "white", "dark", "ticks"
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatterplot with Seaborn Style")
plt.show()
```


---

## ðŸ“Š Basic Plot Types

### 1. **Scatterplot**

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Scatterplot with Categories")
plt.show()
```


---

### 2. **Histogram / Distribution Plot**

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```


---

### 3. **Boxplot**

```python

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Boxplot by Day")
plt.show()
```


---

### 4. **Barplot**

```python
sns.barplot(x="day", y="tip", data=tips, estimator=sum)
plt.title("Total Tips by Day")
plt.show()
```

---

### 5. **Countplot**

```python
sns.countplot(x="day", data=tips)
plt.title("Count of Records by Day")
plt.show()
```


---

## ðŸŒˆ Color Palettes

Seaborn comes with beautiful color palettes:

```python

sns.set_palette("coolwarm")
sns.barplot(x="day", y="tip", data=tips)
plt.title("Coolwarm Palette Example")
plt.show()
```
