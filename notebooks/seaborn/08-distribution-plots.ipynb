# üìä Distribution Plots in Seaborn
**`08-distribution-plots.ipynb`**

Seaborn provides powerful tools for **visualizing distributions** of data.  
These plots are useful for understanding **spread, central tendency, and outliers**.

We‚Äôll use the `tips` dataset for demonstration.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")
tips.head()
```


---

## 1Ô∏è‚É£ Histogram (`histplot`)

```python
sns.histplot(tips["total_bill"], bins=20, kde=False)
plt.title("Histogram of Total Bill")
plt.show()
```


‚úÖ Shows **frequency distribution**.

---

## 2Ô∏è‚É£ Histogram + KDE Overlay

```python

sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Histogram with KDE (Kernel Density Estimate)")
plt.show()
```


‚úÖ KDE curve shows **smoothed distribution**.

---

## 3Ô∏è‚É£ Kernel Density Plot (`kdeplot`)

```python
sns.kdeplot(tips["total_bill"], shade=True)
plt.title("KDE Plot of Total Bill")
plt.show()
```


‚úÖ KDE is a **smooth curve** estimate of the probability distribution.

---

## 4Ô∏è‚É£ Cumulative Distribution

```python
sns.kdeplot(tips["total_bill"], cumulative=True, shade=True)
plt.title("Cumulative Distribution of Total Bill")
plt.show()
```


‚úÖ Helps visualize **proportion of data below a threshold**.

---

## 5Ô∏è‚É£ Rug Plot

```python
sns.rugplot(tips["total_bill"])
plt.title("Rug Plot of Total Bill")
plt.show()
```

‚úÖ Rug plots show **individual observations** along an axis.

---

## 6Ô∏è‚É£ Multiple Distributions

```python
sns.histplot(data=tips, x="total_bill", hue="sex", kde=True, element="step")
plt.title("Distribution by Gender")
plt.show()
```


‚úÖ Compare distributions across categories.

---


## 7Ô∏è‚É£ Joint Distribution (`jointplot`)

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```


‚úÖ Displays **bivariate relationships** with marginal distributions.

Other `kind` options: `"kde"`, `"hex"`, `"reg"`.

---

## 8Ô∏è‚É£ Pairwise Distributions (`pairplot`)


```python

sns.pairplot(tips, hue="sex")
plt.show()
```


‚úÖ Shows **relationships between all numerical variables**.

---

# üìå Summary

- ‚úÖ **`histplot`** ‚Üí histograms with optional KDE  
- ‚úÖ **`kdeplot`** ‚Üí smooth distribution curve  
- ‚úÖ **`rugplot`** ‚Üí raw data points  
- ‚úÖ **`jointplot`** ‚Üí bivariate distribution  
- ‚úÖ **`pairplot`** ‚Üí multiple pairwise distributions
