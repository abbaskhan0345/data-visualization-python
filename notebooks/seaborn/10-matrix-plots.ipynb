Matrix plots are useful for visualizing **relationships between multiple variables** in a grid-like form.  
Seaborn provides **heatmap** and **clustermap** to explore correlations and clustered structures.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
flights = sns.load_dataset("flights")
flights.head()
```


---


## 1Ô∏è‚É£ Heatmap (Basic)

```python
# ‚úÖ Correct pivot
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Flights Heatmap: Passengers per Month-Year")
plt.show()
```


‚úÖ Each cell = number of passengers for that month & year.

## 3Ô∏è‚É£ Masked Heatmap

```python
# Generate mask for upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Masked Correlation Heatmap")
plt.show()
```


‚úÖ Useful to avoid redundancy in symmetric matrices.

---

## 4Ô∏è‚É£ Cluster Map (Hierarchical Clustering)

```python

sns.clustermap(corr, annot=True, cmap="viridis", linewidths=0.5)
```


‚úÖ Automatically **clusters similar variables** together.
