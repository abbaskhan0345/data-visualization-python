
# üìà Regression Plots in Seaborn

Regression plots help visualize the **relationship between two variables** along with the fitted regression line.  
Seaborn provides **`regplot()`** for simple plots and **`lmplot()`** for advanced faceting and grouping.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")
tips.head()
```


---


## 1Ô∏è‚É£ Basic Regression Plot

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Basic Regression Plot: Total Bill vs Tip")
plt.show()
```


‚úÖ Shows scatterplot + linear regression line.

---



## 2Ô∏è‚É£ Regression with Confidence Interval

```python

sns.regplot(x="total_bill", y="tip", data=tips, ci=95)
plt.title("Regression with Confidence Interval (95%)")
plt.show()
```


‚úÖ Shaded region = 95% confidence interval.

---


## 3Ô∏è‚É£ Remove Confidence Interval

```python

sns.regplot(x="total_bill", y="tip", data=tips, ci=None, color="red")
plt.title("Regression without Confidence Interval")
plt.show()
```


---


## 4Ô∏è‚É£ Polynomial Regression

```python

sns.regplot(x="total_bill", y="tip", data=tips, order=2, scatter_kws={"s": 40})
plt.title("Polynomial Regression (Order=2)")
plt.show()
```


‚úÖ Fits a quadratic curve instead of a straight line.

---


## 5Ô∏è‚É£ Logistic Regression



```python
# Example: smoker vs tip > 3 (binary outcome)
tips["big_tip"] = (tips["tip"] > 3).astype(int)

sns.regplot(x="total_bill", y="big_tip", data=tips, logistic=True, ci=None)
plt.title("Logistic Regression: Probability of Big Tip")
plt.show()
```


---


## 6Ô∏è‚É£ lmplot for Grouped Regression

```python

sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips, height=5, aspect=1.2)
plt.title("lmplot: Regression by Sex")
plt.show()
```


‚úÖ Automatically separates regressions by group (`hue`).

---


## 7Ô∏è‚É£ Faceted lmplot

```python
sns.lmplot(x="total_bill", y="tip", col="sex", row="smoker", data=tips, height=4)
```


‚úÖ Creates multiple regression plots across **rows & columns**.

---


## 8Ô∏è‚É£ Style Customization

```python

sns.lmplot(
    x="total_bill", y="tip", hue="day", data=tips,
    palette="Set1", markers=["o", "s", "D", "^"],
    height=5, aspect=1.3
)
plt.title("Custom Styled Regression by Day")
plt.show()
```


---

# üìå Summary

- **regplot()** ‚Üí simple regression (with optional polynomial & logistic fits).  
- **lmplot()** ‚Üí regression with grouping & faceting.  
- **Customization** ‚Üí control scatter style, colors, markers, and layout.
